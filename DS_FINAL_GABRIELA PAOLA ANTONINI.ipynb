#FOOD NUTRITION

##Introducción

El presente proyecto tiene como finalidad desarrollar un modelo de clasificación que permita predecir la cantidad de octógonos de advertencia que cada producto alimenticio debería llevar, basado en su perfil nutricional. Mediante la aplicación de técnicas de aprendizaje automático, se pretende analizar los datos nutricionales para clasificar los alimentos de acuerdo con los criterios de etiquetado definidos en Argentina, brindando así una herramienta útil para la industria alimentaria y los organismos de control.

##Abstract

La creciente preocupación por las enfermedades relacionadas con una alimentación poco saludable ha llevado a la implementación de sistemas de etiquetado en varios países. En Argentina, el etiquetado octogonal advierte a los consumidores sobre el exceso de nutrientes críticos en los alimentos procesados. Este análisis busca desarrollar un modelo de clasificación que prediga la cantidad de octógonos que un producto debería llevar en función de su perfil nutricional. La audiencia potencial que se beneficiaría de este análisis incluye a profesionales de la salud, nutricionistas, reguladores del sector alimentario, y empresas del rubro alimenticio que buscan reformular sus productos para cumplir con las normativas.

##Descripciones de las columnas del Dataset

**Food:** el nombre o tipo de alimento.

**Caloric Value (Valor calórico):** energía total proporcionada por el alimento, generalmente medida en kilocalorías (kcal) por cada 100 gramos.

**Fat (Grasas) (en g):** cantidad total de grasas en gramos por cada 100 gramos, incluidos los desgloses que se indican a continuación.

**Saturated Fats(Grasas saturadas) (en g):**  cantidad de grasas saturadas (grasas que generalmente aumentan el nivel de colesterol en la sangre) en gramos por cada 100 gramos.

**Monounsaturated Fats (Grasas monoinsaturadas) (en g):** cantidad de grasas monoinsaturadas (consideradas grasas saludables para el corazón) en gramos por cada 100 gramos.

**Polyunsaturated Fats (Grasas poliinsaturadas) (en g):** cantidad de grasas poliinsaturadas (incluyen grasas esenciales que el cuerpo necesita pero que no puede producir por sí mismo) en gramos por cada 100 gramos.

**Carbohydrates (Carbohidratos) (en g):** carbohidratos totales en gramos por cada 100 gramos, incluidos los azúcares.

**Sugars (Azúcares) (en g):** azúcares totales en gramos por cada 100 gramos, un subconjunto de los carbohidratos.

**Protein (Proteínas) (en g):** proteínas totales en gramos por cada 100 gramos, esenciales para la reparación y el crecimiento del cuerpo.

**Dietary Fiber (Fibra dietética) (en g):** contenido de fibra en gramos por cada 100 gramos, importante para la salud digestiva.

**Cholesterol (Colesterol) (en mg):** contenido de colesterol en miligramos por cada 100 gramos, pertinente para la salud cardiovascular.

**Sodium (Sodio) (en g):** contenido de sodio en gramos por cada 100 gramos, crucial para el equilibrio de líquidos y la función nerviosa.

**Water (Agua) (en g):** contenido de agua en gramos por cada 100 gramos, que afecta la densidad energética del alimento.

**Vitamina A (en mg):** cantidad de vitamina A en microgramos por cada 100 gramos, importante para la visión y el funcionamiento inmunológico.

**Vitamina B1 (tiamina)(Thiamine) (en mg):** esencial para el metabolismo de la glucosa.

**Vitamina B11 (ácido fólico)(Folic Acid) (en mg):** crucial para la función celular y el crecimiento de los tejidos, particularmente importante durante el embarazo.

**Vitamina B12 (en mg):** importante para el funcionamiento del cerebro y la formación de la sangre.

**Vitamina B2 (riboflavina)(Riboflavin) (en mg):** necesaria para la producción de energía, el funcionamiento de las células y el metabolismo de las grasas.

**Vitamina B3 (niacina) (Niacin) (en mg):** favorece la salud del sistema digestivo, la piel y los nervios.

**Vitamina B5 (ácido pantoténico)(Pantothenic Acid) (en mg):** necesaria para la formación de células sanguíneas y ayuda a convertir los alimentos en energía.

**Vitamina B6 (en mg):** importante para el desarrollo normal del cerebro y para mantener sanos los sistemas nervioso e inmunológico.

**Vitamina C (en mg):** importante para la reparación de todos los tejidos corporales.

**Vitamina D (en mg):** crucial para la absorción de calcio, promoviendo el crecimiento y la salud de los huesos.

**Vitamina E (en mg):** actúa como antioxidante, ayudando a proteger las células del daño causado por los radicales libres.

**Vitamina K (en mg):** necesaria para la coagulación de la sangre y la salud de los huesos.

**Calcium (Calcio) (en mg):** vital para la formación y el mantenimiento de huesos y dientes fuertes.

**Copper (Cobre) (en mg):** ayuda a la formación de colágeno, aumenta la absorción de hierro y desempeña un papel en la producción de energía.

**Iron (Hierro) (en mg):** esencial para la creación de glóbulos rojos.

**Magnesium (Magnesio) (en mg):** importante para muchos procesos del cuerpo,
incluida la regulación de la función muscular y nerviosa, los niveles de azúcar en sangre y la presión arterial, y la producción de proteínas, huesos y ADN.

**Manganese (Manganeso) (en mg):** interviene en la formación de huesos, factores de coagulación sanguínea y enzimas que desempeñan un papel en el metabolismo de grasas y carbohidratos, la absorción de calcio y la regulación del azúcar en sangre.

**Phosphorus (Fósforo) (en mg):** ayuda a la formación de huesos y dientes y es necesario para que el cuerpo produzca proteínas para el crecimiento, mantenimiento y reparación de células y tejidos.

**Potassium (Potasio) (en mg):** ayuda a regular el equilibrio de líquidos, las contracciones musculares y las señales nerviosas.

**Selenium (Selenio) (en mg):** importante para la reproducción, la función de la glándula tiroides, la producción de ADN y la protección del cuerpo contra el daño causado por los radicales libres y las infecciones.

**Zinc (en mg):** necesario para el correcto funcionamiento del sistema inmunitario y desempeña un papel en la división celular, el crecimiento celular, la cicatrización de heridas y la descomposición de los carbohidratos.

**Nutrition Density (Densidad nutricional):** métrica que indica la riqueza nutricional de los alimentos por caloría.

```python
# Obtengo las estadísticas descriptivas
df_stats = df.describe().round(5)

# Transpongo para que las columnas sean las estadísticas
df_stats_transpuesta = df_stats.T
df_stats_transpuesta['coeficiente_de_variacion'] = (df_stats_transpuesta['std'] / df_stats_transpuesta['mean']) * 100
df_stats_transpuesta['IQR']=(df_stats_transpuesta['75%'] - df_stats_transpuesta['25%'])
df_stats_transpuesta['ceros'] = (df.select_dtypes('number')==0).sum()
df_stats_transpuesta['porcentaje_ceros'] = (df_stats_transpuesta['ceros'] / df.shape[0]) * 100
# Mostrar el DataFrame de estadísticas
#print(df_stats_transpuesta)# Reiniciar el índice y establecer la primera columna como nombre
df_stats_transpuesta.reset_index(inplace=True)
df_stats_transpuesta.rename(columns={'index': 'Nutrientes'}, inplace=True)


from IPython.display import display
display(df_stats_transpuesta)
```

- Alto porcentaje de ceros

 Algunos nutrientes como la **Vitamina A,D, E y K** tienen un porcentaje muy alto de ceros (> 40%), lo cual podría influir en los análisis si los valores faltantes son importantes en este caso considero que  simplemente significa que la mayoría de los alimentos no poseen estas vitaminas.

- Alta variabilidad

 El coeficiente de variación es muy alto en nutrientes como el **colesterol y la vitamina C**, lo que indica que los valores están muy dispersos en torno a la media.

- Valores extremos

 Algunas variables como el **colesterol, vitamina A y selenio** tienen valores máximos significativamente más altos que la media, lo que puede indicar la presencia de outliers.

- IQR bajo

 En algunas vitaminas como la **Vitamina K**, el rango intercuartílico es bajo, lo que sugiere que la mayoría de los alimentos tienen niveles similares de ese nutriente.

```python
plt.figure(figsize=(10,6))
plt.boxplot(df_stats_transpuesta['porcentaje_ceros'], vert=False, patch_artist=True)
plt.xlabel('Porcentaje de Ceros')
plt.title('Boxplot del Porcentaje de Ceros')
plt.show()
```

```python
# Cre0 una figura con 3 subplots, uno al lado del otro en una columna
fig, axes = plt.subplots(1, 3, figsize=(20,8))

# Primer gráfico: Porcentaje de Ceros por Nutriente
axes[0].barh(df_stats_transpuesta['Nutrientes'], df_stats_transpuesta['porcentaje_ceros'], color='skyblue')
axes[0].set_xlabel('Porcentaje de Ceros')
axes[0].set_title('Porcentaje de Ceros por Nutriente')

# Segundo gráfico: Rango Intercuartil (IQR) por Nutriente
axes[1].barh(df_stats_transpuesta['Nutrientes'], df_stats_transpuesta['IQR'], color='skyblue')
axes[1].set_xlabel('Rango Intercuartil (IQR)')
axes[1].set_title('Rango Intercuartil por Nutriente')

# Tercer gráfico: Coeficiente de Variación por Nutriente
axes[2].barh(df_stats_transpuesta['Nutrientes'], df_stats_transpuesta['coeficiente_de_variacion'], color='skyblue')
axes[2].set_xlabel('Coeficiente de Variación')
axes[2].set_title('Coeficiente de Variación por Nutriente')

# Ajustar el layout para evitar que los gráficos se superpongan
plt.tight_layout()

# Mostrar los gráficos juntos
plt.show()
```

###Algoritmos de Clasificación

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split,cross_val_score
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report
from sklearn.feature_selection import SelectKBest, mutual_info_classif
from sklearn.preprocessing import StandardScaler
# Importar algunos modelos de ejemplo
from sklearn.linear_model import LogisticRegression
from sklearn.multiclass import OneVsRestClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from tabulate import tabulate


# Función para evaluar múltiples modelos
def evaluate_models(models, X, y, test_size=0.2, random_state=42,cv=3):
    results={}
    conf_matrices = {}  # Diccionario para almacenar matrices de confusión

    # Dividir los datos en conjunto de entrenamiento y prueba
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=random_state)

     # Escalar los datos
    scaler = StandardScaler()
    X_train_scaled = scaler.fit_transform(X_train)
    X_test_scaled = scaler.transform(X_test)


    # Selección de características
    selector = SelectKBest(score_func=mutual_info_classif, k=10)  # Selecciona las 10 mejores características
    X_train_selected = selector.fit_transform(X_train, y_train)
    X_test_selected = selector.transform(X_test)

    # Evaluar cada modelo
    for model_name, model in models.items():
       # Validación cruzada
        cv_scores = cross_val_score(model, X_train_selected, y_train, cv=cv)
        # Entrenar el modelo
        model.fit(X_train_selected, y_train)

        # Predecir en el conjunto de prueba
        y_pred = model.predict(X_test_selected)


        # Agregar el classification report
        report = classification_report(y_test, y_pred, zero_division=1, output_dict=True)
       # Extraer métricas relevantes para cada clase y las métricas globales
        metrics = {}
        for class_label, metrics_data in report.items():
            if class_label not in ['accuracy', 'macro avg', 'weighted avg']:
                metrics[class_label] = {
                    'Precision': metrics_data['precision'],
                    'Recall': metrics_data['recall'],
                    'F1 Score': metrics_data['f1-score'],
                    'Support': metrics_data['support']
                }

        # Agregar también las métricas promedio ponderadas
        metrics['Weighted Avg'] = {
            'Precision': report['weighted avg']['precision'],
            'Recall': report['weighted avg']['recall'],
            'F1 Score': report['weighted avg']['f1-score'],
            'Support': report['weighted avg']['support']
        }

        results[model_name] = metrics
        results[model_name]['Cross-Validation Mean Score'] = np.mean(cv_scores)
        results[model_name]['Cross-Validation Std Score'] = np.std(cv_scores)

        # Calcular matriz de confusión
        conf_matrices[model_name] = confusion_matrix(y_test, y_pred)



    return results, conf_matrices



if __name__ == "__main__":

    X = df_new.drop(columns=['Octogonos'])  # Características
    y = df_new['Octogonos']  # Variable objetivo


    # Definir modelos
    models = {

        'Regresion Logistica': OneVsRestClassifier(LogisticRegression(class_weight='balanced',max_iter=1000, C=0.5, solver='liblinear')),
        'KNN': KNeighborsClassifier(n_neighbors=5),
        'SVC': SVC(kernel='linear'),
        'Random Forest': RandomForestClassifier(random_state=42)
    }



    # Evaluar modelos
    results,conf_matrices = evaluate_models(models, X, y)


    # Convertir resultados a DataFrame y mostrar en formato tabulado
    for model_name, metrics in results.items():
        metrics_df = pd.DataFrame(metrics).T
        metrics_df.index.name = 'Clase'
        print(f"\nResultados para {model_name}:\n")
        print(tabulate(metrics_df, headers='keys', tablefmt='grid'))
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report
from sklearn.feature_selection import SelectKBest, mutual_info_classif
from sklearn.preprocessing import StandardScaler
from imblearn.over_sampling import SMOTE  # Importar SMOTE
# Importar algunos modelos de ejemplo
from sklearn.linear_model import LogisticRegression
from sklearn.multiclass import OneVsRestClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from tabulate import tabulate

# Función para evaluar múltiples modelos
def evaluate_models_opt(models, X, y, test_size=0.2, random_state=42, cv=3):
    results = {}
    conf_matrices_opt = {}  # Diccionario para almacenar matrices de confusión

    # Dividir los datos en conjunto de entrenamiento y prueba
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=random_state)

    # Escalar los datos
    scaler = StandardScaler()
    X_train_scaled = scaler.fit_transform(X_train)
    X_test_scaled = scaler.transform(X_test)

    # Selección de características
    selector = SelectKBest(score_func=mutual_info_classif, k=10)  # Selecciona las 10 mejores características
    X_train_selected = selector.fit_transform(X_train_scaled, y_train)
    X_test_selected = selector.transform(X_test_scaled)

    # Aplicar SMOTE solo al conjunto de entrenamiento
    smote = SMOTE(sampling_strategy='auto', k_neighbors=2, random_state=42)
    X_train_resampled, y_train_resampled = smote.fit_resample(X_train_selected, y_train)

    # Evaluar cada modelo
    for model_name, model in models.items():
        # Validación cruzada
        cv_scores = cross_val_score(model, X_train_resampled, y_train_resampled, cv=cv)
        # Entrenar el modelo
        model.fit(X_train_resampled, y_train_resampled)

        # Predecir en el conjunto de prueba
        y_pred = model.predict(X_test_selected)

        # Agregar el classification report
        report = classification_report(y_test, y_pred, zero_division=1, output_dict=True)
        # Extraer métricas relevantes para cada clase y las métricas globales
        metrics = {}
        for class_label, metrics_data in report.items():
            if class_label not in ['accuracy', 'macro avg', 'weighted avg']:
                metrics[class_label] = {
                    'Precision': metrics_data['precision'],
                    'Recall': metrics_data['recall'],
                    'F1 Score': metrics_data['f1-score'],
                    'Support': metrics_data['support']
                }

        # Agregar también las métricas promedio ponderadas
        metrics['Weighted Avg'] = {
            'Precision': report['weighted avg']['precision'],
            'Recall': report['weighted avg']['recall'],
            'F1 Score': report['weighted avg']['f1-score'],
            'Support': report['weighted avg']['support']
        }

        results[model_name] = metrics
        results[model_name]['Cross-Validation Mean Score'] = np.mean(cv_scores)
        results[model_name]['Cross-Validation Std Score'] = np.std(cv_scores)

        # Calcular matriz de confusión
        conf_matrices_opt[model_name] = confusion_matrix(y_test, y_pred)

    return results, conf_matrices_opt


if __name__ == "__main__":
    X = df_new.drop(columns=['Octogonos'])  # Características
    y = df_new['Octogonos']  # Variable objetivo

    # Definir modelos
    models = {
        'Regresion Logistica': OneVsRestClassifier(LogisticRegression(class_weight='balanced', max_iter=1000, C=0.5, solver='liblinear')),
        'KNN': KNeighborsClassifier(n_neighbors=5),
        'SVC': SVC(kernel='linear'),
        'Random Forest': RandomForestClassifier(random_state=42)
    }

    # Evaluar modelos
    results, conf_matrices_opt = evaluate_models_opt(models, X, y)

    # Convertir resultados a DataFrame y mostrar en formato tabulado
    for model_name, metrics in results.items():
        metrics_df = pd.DataFrame(metrics).T
        metrics_df.index.name = 'Clase'
        print(f"\nResultados para {model_name}:\n")
        print(tabulate(metrics_df, headers='keys', tablefmt='grid'))
```
